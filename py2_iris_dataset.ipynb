## $y = a_1X_1 + a_2X_2 + a_3X_3 + a_4X_4 + e$

De dataset wordt gesplits in de X'en en de y:

- de X'en zijn alle kolommen (features) minus de "doelvariabele" y
- de y is de doelvariabele of target (1 kolom)

```python
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
```

Voor toelichting zie [sklearn.cross_validation](https://scikit-learn.org/stable/modules/cross_validation.html)

## Bayes, Thomas (1701-1761)

```python
# Initial model: GaussianNB
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                   # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
# Alternate model: MultinomialNB
from sklearn.naive_bayes import MultinomialNB # 1. choose model class
model = MultinomialNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                      # 3. fit model to data
y_model = model.predict(Xtest)                # 4. predict on new data
accuracy_score(ytest, y_model)
```

```python
# Alternate model: ComplementNB
from sklearn.naive_bayes import ComplementNB # 1. choose model class
model = ComplementNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                     # 3. fit model to data
y_model = model.predict(Xtest)               # 4. predict on new data
accuracy_score(ytest, y_model)
```

```python
# Alternate model: BernoulliNB
from sklearn.naive_bayes import BernoulliNB # 1. choose model class
model = BernoulliNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                    # 3. fit model to data
y_model = model.predict(Xtest)              # 4. predict on new data
accuracy_score(ytest, y_model)
```
