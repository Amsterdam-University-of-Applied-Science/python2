```python
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

x = np.arange(10)
y = np.sin(x)

plt.xkcd()

fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y)
#ax.text(2, 2, 'Voorbeeld van een grafiek', size=16)

plt.title('Y=sin(x)')

plt.show()
```
