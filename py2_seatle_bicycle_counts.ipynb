```python
# load the data into a pandas dataframe
import pandas as pd

data = pd.read_csv('FremontBridge.csv', index_col='Date', parse_dates=True)
```

```python
# display the data
data.head()
```

```python
# shorten column names and add total
data.columns = ['Total','West', 'East']
```

```python
# display the data again
data.head()
```

```python
# summary statistics for this data
data.dropna().describe()
```

### Visualising the data

```python
# loading visualisation libraries
%matplotlib inline
data.plot()
```

```python
# resampling the data and plot again
weekly = data.resample('W').sum()
weekly.plot(style=[':', '--', '-'])
```

```python
# add rolling mean
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'])
```

```python
# smooth the rolling mean
daily.rolling(50, center=True,win_type='gaussian').sum(std=10).plot(style=[':', '--', '-']);
```

### Digging into the data

```python
# look at average daily traffic by hour
# transform data by using of groupby
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-']);
```

```python
# alternatively: look at daily traffic by day of the week
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-']);
```

```python
# now compare hourly trend of weekdays with hourly trend of weekends
weekend = np.where(data.index.weekday < 5, 'Weekday', 'Weekend')
by_time = data.groupby([weekend, data.index.time]).mean()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
by_time.loc['Weekday'].plot(ax=ax[0], title='Weekdays',
                           xticks=hourly_ticks, style=[':', '--', '-'])
by_time.loc['Weekend'].plot(ax=ax[1], title='Weekends',
                           xticks=hourly_ticks, style=[':', '--', '-']);
```
